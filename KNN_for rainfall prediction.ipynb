```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillUp/labs/ML-FinalAssignment/Weather_Data.csv')
```

```python
df=df.drop(['Date','WindGustDir','WindDir9am','WindDir3pm'],axis =1)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
cols=['RainToday','RainTomorrow']
df[cols]=df[cols].apply(LabelEncoder().fit_transform)
```

```python
X=df[['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine',
       'WindGustSpeed', 'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am',
       'Humidity3pm', 'Pressure9am', 'Pressure3pm', 'Cloud9am', 'Cloud3pm',
       'Temp9am', 'Temp3pm', 'RainToday']].values
```

```python
X[0:5]
```

```python
y=df['RainTomorrow']
```

```python
y[0:5]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=4)
```

```python
print("Training shape = ", X_train.shape)
print("Testing shape = ", X_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
k=4
neigh=KNeighborsClassifier(n_neighbors=k).fit(X_train,y_train)
neigh
```

```python
y_hat=neigh.predict(X_test)
```

```python
y_hat[0:5]
```

```python
y_test[0:5]
```

```python
from sklearn import metrics
```

```python
print("The accuracy of the training is ", metrics.accuracy_score(y_train,neigh.predict(X_train)))
```

```python
print("The accuracy of the testing is ",metrics.accuracy_score(y_hat,y_test))
```

```python
accuracy=[]
for i in range(1,11):
    neigh=KNeighborsClassifier(n_neighbors=i).fit(X_train,y_train)
    y_hat=neigh.predict(X_test)
    accuracy.append(metrics.accuracy_score(y_hat,y_test))
```

```python
print(accuracy)
```

```python
x=max(accuracy)
k=1
for i in accuracy:
    if(i==x):
        break
    else:
        k=k+1
print('the max accuracy is ',x ,'for k = ',k)
    
```
