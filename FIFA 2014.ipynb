# Importing libraries 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Loading Dataset

```python
df=pd.read_csv('WorldCupMatches.csv')
```

```python
df
```

# Data Preprocessing 

```python
df.isnull().sum()
```

```python
df=df.drop(['Year','Datetime','Stage','Stadium','City','Referee'],axis=1)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
for i in df.columns:
    df[i]=le.fit_transform(df[i])
```

```python
df.head()
```

```python
df.isnull().sum()
```

# Loading the second Dataset 

```python
df1=pd.read_csv('WorldCupPlayers.csv')
```

```python
df1
```

# Data preprocessing

```python
df1=df1.drop(['Coach Name','Shirt Number','Event'],axis=1)
```

```python
df1
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
for i in df1.columns:
    df1[i]=le.fit_transform(df1[i])
```

```python
df1.isnull().sum()
```

# Loading The third Dataset

```python
df2=pd.read_csv('WorldCups.csv')
```

```python
df2
```

# Data Preprocessing 

```python
for i in df2.columns:
    df2[i]=le.fit_transform(df2[i])
```

```python
df2
```

```python
df2.isnull().sum()
```

# Margeing The three Datasets

```python
df3=pd.merge(df,df1)
```

```python
data=pd.merge(df2,df3)
```

```python
data
```

```python
data.columns
```

```python
X=data[['Year', 'Country', 'Runners-Up', 'Third', 'Fourth',
       'GoalsScored', 'QualifiedTeams', 'MatchesPlayed', 'Attendance',
       'Home Team Name', 'Home Team Goals', 'Away Team Goals',
       'Away Team Name', 'Win conditions', 'Half-time Home Goals',
       'Half-time Away Goals', 'Assistant 1', 'Assistant 2', 'RoundID',
       'MatchID', 'Home Team Initials', 'Away Team Initials', 'Team Initials',
       'Line-up', 'Player Name', 'Position']]
```

```python
y=data['Winner']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
print('The shape of training is ',X_train.shape)
print('The shape of testing is ',y_test.shape)
```

# Model Building Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lo=LogisticRegression()
```

```python
lo.fit(X_train,y_train)
```

```python
y_hat=lo.predict(X_test)
```

```python
y_hat[0:5]
```

```python
y_test[0:5]
```

```python
from sklearn import metrics 
```

```python
print('The assuracy of the model is ',metrics.accuracy_score(y_hat,y_test))
```

 # Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree=DecisionTreeClassifier(criterion='entropy')
```

```python
tree.fit(X_train,y_train)
```

```python
y_pred=tree.predict(X_test)
```

```python
y_pred[0:20]
```
